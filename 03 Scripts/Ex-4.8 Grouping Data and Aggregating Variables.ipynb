### Q 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Aggregatig Data with Transform()
df_ords_prods_merge['mean_prices'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merge.head(10)
```

### Applying the various conditions for the Spending flag

```python
# Creating a Spending flag by using loc() by assigpning into a new column(mean_prices)
df_ords_prods_merge.loc[df_ords_prods_merge['mean_prices'] < 10,'Spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_prices'] >= 10,'Spending_flag'] = 'High Spender'
```

### Finding the frequency of the Spending flag

```python
# Finding the frequency of Spending_flag
df_ords_prods_merge['Spending_flag'].value_counts(dropna=False)
```

### Q 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Aggregatig Data- finding median with Transform()
df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Testing the dataframe
df_ords_prods_merge.tail(10)
```

### Applying the conditions for frquency flag

```python
 # Creating a order frequency flag by using loc() by assigpning into a new column(order_frequency)
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency']) > 20,'Frequency_flag'] = 'Non Frequent Customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] <= 20) & (df_ords_prods_merge['order_frequency'] > 10), 'Frequency_flag'] = 'Regular Customer' 
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency']) <= 10,'Frequency_flag'] = 'Frequent Customer'
```

### Calculating the frequency of frquency flag

```python
# Finding the frequency of Frequency_flag
df_ords_prods_merge['Frequency_flag'].value_counts(dropna=False)
```
