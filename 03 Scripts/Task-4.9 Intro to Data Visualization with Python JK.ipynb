### Plotting bar charts for categorical data

```python
# Plotting the bar graph 
df_ords_prods_merge['order_day_of_week'].value_counts().plot.bar()
plt.xlabel('Days of the week')
plt.ylabel('Number of orders')
plt.title('Busiest days of the week')
```

```python
# Indexing the bar graph 
df_ords_prods_merge['order_day_of_week'].value_counts().sort_index().plot.bar()
plt.xlabel('Days of the week')
plt.ylabel('Number of orders')
plt.title('Busiest days of the week')
```

```python
# Saving the bar graph to bar object
bar = df_ords_prods_merge['order_day_of_week'].value_counts().plot.bar()
plt.xlabel('Days of the week')
plt.ylabel('Number of orders')
plt.title('Busiest days of the week')
```

```python
# Saving the bar chart 
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```
